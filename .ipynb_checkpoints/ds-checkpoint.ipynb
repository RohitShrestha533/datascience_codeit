```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'student_dataset.csv')
data
```

```python
data.placement_status.value_counts()
```

```python
plt.title("status count of students")
plt.xlabel("Placement type")
plt.ylabel("number of students")
sns.countplot(x='placement_status',hue='placement_status',data=data)
plt.tight_layout()
plt.savefig("placementBar.jpg")
```

```python
sns.lineplot(x='placement_status',y='study_hours',data=data)
```
