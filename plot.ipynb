```python
sns.barplot(x='gender',y='daily_screen_hours',data=data)
```

```python
data
```

```python
#line chart
sns.lineplot(x='occupation',y='daily_screen_hours',data=data)
```

```python
# scatterplot
sns.scatterplot(x='occupation',y='daily_screen_hours',data=data)
```

```python
# KDE PLOT  kenral density estimatro
sns.kdeplot(x='daily_screen_hours',data=data)
plt.axvline(data['daily_screen_hours'].mean(),linestyle='--',color='red')
plt.axhline(0.02,linestyle='--')
```

```python

sns.kdeplot(x='daily_screen_hours',data=data)
plt.axvline(data['daily_screen_hours'].mean(),linestyle='--',color='red')
```

```python
# histogram
sns.histplot(x='daily_screen_hours',data=data)
```

```python
plt.figure(figsize=(20,5))
sns.histplot(x='daily_screen_hours',kde=True,data=data)
plt.xticks(rotation=90,fontsize=12)
plt.show()
```

```python
# box plot
sns.boxplot(x='occupation',y='daily_screen_hours',data=data)
```

```python
sns.violinplot(x='occupation',y='daily_screen_hours',data=data)
```

```python
plt.figure(figsize=(20,5))
sns.violinplot(x='occupation',y='daily_screen_hours',inner='quart',hue='gender',data=data)
```
