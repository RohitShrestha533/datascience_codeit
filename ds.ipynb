```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'social_media_screentime_mental_health_2026.csv')
data
```

```python
data.occupation.value_counts()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
plt.title("occupation's counts")
plt.xlabel("Placement type")
plt.ylabel("number of people")
sns.countplot(x='occupation',data=data)
plt.xticks(rotation=75)
plt.tight_layout()
plt.savefig("occupation.jpg")
```

```python
sns.lineplot(x='occupation',y='daily_screen_hours',hue='gender',ci=None,linestyle='--',color='green',marker='D',data=data)
avg = data['daily_screen_hours'].mean()
plt.xlabel("occupations")
plt.ylabel("daily screen time in hrs")
plt.xticks(rotation=75)
plt.axhline(avg,color='red')
plt.show()
```

# dataframe merge and concatenate

```python
x = list(range(9,99,9))
y = list(range(11,111,11))
z = list(range(14,144,14))
data_1 = pd.DataFrame({
    'X':x,
    'Y':y,
    'Z':z
})
# data_1['index1']=range(len(data_1))
data_1
```

```python
a = list(range(7,77,7))
b = list(range(13,133,13))
c = list(range(6,66,6))
data_2 = pd.DataFrame({
    'A':a,
    'B':b,
    'C':c
})
# data_2['index2']=range(len(data_2))
data_2
```

```python
data_1.head(2)
```

```python
data_2.head(2)
```

```python
pd.concat([data_1,data_2])
```

```python
pd.merge(data_1,data_2,left_index=True,right_index=True)
```
