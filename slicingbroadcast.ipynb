```python
x = ['a','b','c','d','e','m','n']
y = ['f','g','h','i','j','q','r']
z = ['k','l','m','n','o','s','t']
g = ['p','q','r','s','t','q','r']
h = ['p','q','r','s','t','h','i']
i = ['p','q','r','s','t','b','c']
```

```python
array = np.array([x,y,z,g,h,i])
array
```

```python
# select row from 1 to 3
array[1:3]
```

```python
# select col from index 2 to 4
array[:,2:4]
```

```python
# select from row 1 to 3 and col 4 to 7
array[1:3,4:7]
```

maths operations

```python
a2 =  np.array([range(1,5),range(2,6),range(3,7),range(4,8),range(5,9)])
a2
```

```python
np.sum(a2) # sum of all
```

```python
np.sum(a2,axis=0) # sum of each col 
```

```python
np.sum(a2,axis=1) # sum of each row
```

```python
np.mean(a2)
```

```python
np.mean(a2,axis=0) # mean of col
```

```python
np.mean(a2,axis=1) # mean of row 
```

```python
np.max(a2)
```

```python
np.max(a2,axis=0) # max number's row
```

```python
np.max(a2,axis=1) # max number col
```

```python
np.min(a2)
```

```python
np.min(a2,axis=0)
```

```python
np.min(a2,axis=1)
```
